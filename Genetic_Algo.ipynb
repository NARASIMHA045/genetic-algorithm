```python
import random
import string
```

```python
target = "my name is swamy"
population_size = 100
mutation_rate = 0.01
genes = string.ascii_lowercase + " "
```

```python

# we create random childen for the evolution
def create():
  return "".join(random.choice(genes) for z in range(len(target)))
```

```python

# calculate the fitness, return the sum at which the each char matches the target
def fitness(s):
  return sum(1 for a,b in zip(s,target) if a== b)
  #  return sum( 1 for i in range(len(target))
  #   if(s[i]==target[i]))
```

```python

# we create child by crossover of 2 parents a and b
def crossover(a,b):
  p=random.randint(0,len(target)-1)
  return a[:p] + b[p:]
```

```python

# Mutation, explore new possible solutions
def mutate(s):
  result= ""
  for c in s:
    if random.random() < mutation_rate:
      result+=random.choice(genes)
    else:
      result+=c
  return result
```

```python
population = [create() for _ in range(population_size)]
```

```python
for generation in range(1000):
  population.sort(key=fitness,reverse=True)
  best=population[0]
  print(generation,"->",best)
  # found the target
  if best == target:
    print("Done!")
    break

  new_pop = population[:10]

  while len(new_pop)<population_size:
    # create parents
    par1=random.choice(population[:50])
    par2=random.choice(population[:50])
    # create new children from the best parents
    child = crossover(par1,par2)
    child=mutate(child)
    new_pop.append(child)
  population = new_pop

```
